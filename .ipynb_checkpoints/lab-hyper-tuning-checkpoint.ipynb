```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship
```

```python
spaceship.isnull().sum().sort_values(ascending=False)
spaceship = spaceship.dropna()
spaceship.drop(columns = ["PassengerId", "Name"],inplace = True)
```

```python
spaceship['Cabin'] = spaceship['Cabin'].apply(lambda x: x[0])
```

```python
spaceship 
```

```python
spaceship.dtypes
```

```python
spaceship.columns 
```

```python
features = spaceship.drop(columns = ["Transported"]) 
target = spaceship["Transported"]
```

```python
features = pd.get_dummies(features, columns=['HomePlanet', 'CryoSleep', 'Cabin', 'Destination','VIP'], drop_first=True)
```

**Perform Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=42)
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


### Scaling

```python
#your code here
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### KNN Classifier without scaling

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
#your code here
log_reg.score(X_test, y_test)
```

```python
X_test_scaled.shape
```

```python
y_pred = log_reg.predict(X_test_scaled)
```

```python
y_pred
```

#### Recall

```python
from sklearn.metrics import recall_score
```

```python
recall_score(y_test, y_pred)
```

#### confusion matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
log_reg.intercept_
```

```python
log_reg.coef_
```


### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth = 3)
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(15, 10))
plot_tree(dt, feature_names=X_train.columns, filled=True)
plt.show() 
```

### Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
recall_score(y_test, y_pred)
```
